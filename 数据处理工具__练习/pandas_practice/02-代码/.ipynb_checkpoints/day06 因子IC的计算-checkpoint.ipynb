# alphalens计算因子的IC

```python
#1. 准备factor
#2. 准备price
#3. 获取计算IC值需要的数据格式，multiIndex的series
#4. 计算IC
```

```python
import alphalens
```

### 1. 准备factor

```python

START_DATE = "20100101"
END_DATE = "20110101"
C27LIST = industry("C27")

trading_dates = get_trading_dates(start_date=START_DATE,end_date=END_DATE)

q = query(fundamentals.financial_indicator.earnings_per_share).filter(fundamentals.stockcode.in_(C27LIST))
fund = get_fundamentals(q,entry_date=END_DATE,interval="{}d".format(len(trading_dates)))
```

```python
factor = fund["earnings_per_share",:,:].stack().sort_index(ascending=False)
```

```python
factor
```

### 2. 准备price

```python
prices = get_price(C27LIST,start_date=START_DATE,end_date=END_DATE,fields="close")
```

```python
prices
```

### 3. 获取计算IC值需要的数据格式，multiIndex的series

```python
factor_data = alphalens.utils.get_clean_factor_and_forward_returns(factor,prices)
```

```python
factor_data
```

### 4. 计算IC

```python
IC = alphalens.performance.factor_information_coefficient(factor_data)
```

```python
IC
```

# IC的绘制

```python
# 绘制直方图
alphalens.plotting.plot_ic_hist(IC)
```

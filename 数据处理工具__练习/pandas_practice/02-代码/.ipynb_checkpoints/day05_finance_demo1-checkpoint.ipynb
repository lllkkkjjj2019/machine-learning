### 1 绘制日K线图

```python
import pandas as pd
from mpl_finance import candlestick_ochl
import matplotlib.pyplot as plt
from matplotlib.pylab import date2num
```

```python
help(candlestick_ochl)
```

```python
# 读取股票数据
data = pd.read_csv("./stock_day/stock_day.csv")
```

```python
# 调整股票时间顺序
data = data.sort_index()
```

```python
# 将字符串类型的日期——时间类型
data.index = pd.to_datetime(data.index)
```

```python
data.index
```

```python
# 将时间类型转换成浮点型
data["time"] = date2num(data.index)
```

```python
data.head()
```

```python
# 准备ax
figure, axes = plt.subplots(nrows=1, ncols=1, figsize=(13, 8), dpi=100)

# 准备quotes
day_k = data[["time", "open", "close", "high", "low"]]

# 绘制日K线图
candlestick_ochl(axes, day_k.values[:30], width=0.4, colorup="r", colordown="g")

# 显示日期
axes.xaxis_date()

# 显示网格
plt.grid(linestyle="--", alpha=0.5)

# 显示图像
plt.show()
```

### 2 绘制周K线图

```python
# 准备好周K数据
data2 = data.resample("W").last()
```

```python
data
```

```python
# 准备开盘价
data2["open"] = data["open"].resample("W").first()
```

```python
# 准备最高价
data2["high"] = data["high"].resample("W").max()
```

```python
# 准备最低价
data2["low"] = data["low"].resample("W").low()
```

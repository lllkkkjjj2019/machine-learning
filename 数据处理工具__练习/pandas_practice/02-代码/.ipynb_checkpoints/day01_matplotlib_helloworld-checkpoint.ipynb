### 2.2.2 完善原始折线图1(辅助显示层)

```python
# 需求：画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度
# 1 准备数据并画出初始折线图
import random
# 1、准备数据
x = range(60)
y = [random.uniform(15, 18) for i in x]

# 2、创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3、绘制折线图
plt.plot(x, y)

# 4、显示图像
plt.show()
```

```python
# 2 添加自定义x,y刻度
import random
# 1、准备数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]

# 2、创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3、绘制折线图
plt.plot(x, y_shanghai)

# 添加x、y的刻度
# 准备x轴刻度字符串
x_shanghai_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5], x_shanghai_label[::5])
plt.yticks(range(35)[::5])

# 添加网格
plt.grid(True, linestyle="--", alpha=0.5)

# 添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("中午11点到12点温度变化状况")

# 4、显示图像
plt.show()
```

```python
# 2 完善原始折线图2(图像层)
# 收集到北京当天温度变化情况，温度在1度到3度。 
import random
# 1、准备数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 3) for i in x]

# 2、创建画布
plt.figure(figsize=(20, 8), dpi=80)

# 3、绘制折线图
plt.plot(x, y_shanghai, color="m", linestyle="-.", label="上海")
plt.plot(x, y_beijing, color="g", label="北京")

# 显示图例
plt.legend()

# 添加x、y的刻度
# 准备x轴刻度字符串
x_shanghai_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5], x_shanghai_label[::5])
plt.yticks(range(35)[::5])

# 添加网格
plt.grid(True, linestyle="--", alpha=0.5)

# 添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("中午11点到12点温度变化状况")

# 4、显示图像
plt.show()
```
